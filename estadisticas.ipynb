```python
# se convierte en una variable categorica ya que en type, es la unica variable tipo object
df["Target"] = df["Target"].astype('category')

df["Target"] = df["Target"].cat.codes

#verificar si columna cambio
columna = df['Target']
columna.head(20)
```

```python
#filtro de  X con las variables independientes

X = df[['Marital status', 'Application mode', 'Application order', 'Course',
       'Daytime/evening attendance', 'Previous qualification', 'Nacionality',
       "Mother's qualification", "Father's qualification",
       "Mother's occupation", "Father's occupation", 'Displaced',
       'Educational special needs', 'Debtor', 'Tuition fees up to date',
       'Gender', 'Scholarship holder', 'Age at enrollment', 'International',
       'Curricular units 1st sem (credited)',
       'Curricular units 1st sem (enrolled)',
       'Curricular units 1st sem (evaluations)',
       'Curricular units 1st sem (approved)',
       'Curricular units 1st sem (grade)',
       'Curricular units 1st sem (without evaluations)',
       'Curricular units 2nd sem (credited)',
       'Curricular units 2nd sem (enrolled)',
       'Curricular units 2nd sem (evaluations)',
       'Curricular units 2nd sem (approved)',
       'Curricular units 2nd sem (grade)',
       'Curricular units 2nd sem (without evaluations)', 'Unemployment rate',
       'Inflation rate', 'GDP']]

X.head()
```

```python
#filtro de  Y con las variables dependientes
Y = df[['Target']]
Y
```

```python
#chequeo y balanceo

import seaborn as sns

print(Y.value_counts())

# es un balanceo desbalanceado entonces se debe hacer el metodo smote 
```

```python
#funcion para balancear
from imblearn.over_sampling import SMOTE

smote = SMOTE()
x_smote,y_smote = smote.fit_resample(X,Y)
```

```python
import seaborn as sns

print(y_smote.value_counts())
```

```python
#set de datos entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_smote,y_smote,test_size=0.2)
```

```python
#escalar variables ya que los datos escalan diferentes

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#modelo de regresion logistica 
from sklearn.linear_model import LogisticRegression
regresion_log = LogisticRegression()
regresion_log.fit(X_train,y_train)
```

```python
#prediccion de datos de X_test
y_pred = regresion_log.predict(X_test)
y_pred
```

```python
# matriz de confusion
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test,y_pred))
```
